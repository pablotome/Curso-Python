```python
# se declara la función con un hint del tipo de dato esperado en el parámetro valores
# la validación hay que realizarla igual, Python no valida los tipos indicados,
# (son hints)
def multi_pos_pares(valores: list):
  total = 1

  # se valida haber recibido una lista
  if type(valores) == list:

    # se itera por los valores pares usando slicing ::2
    for value in valores[::2]:

      # se valida el valor 
      if type(value) == int or type(value) == float:

        # calculos
        total *= value

      else:
        print("Se descarta valor {}".format(value))

    return total

  else:
    print("Se espera una lista")


# Opción 2   
def multi_pos_pares_v2(valores: list):
  
  if type(valores) == list:
    total = 1

    # se crea una función lambda para usar como condición en 
    es_int_o_float = lambda x: type(x) == int or type(x) == float

    # se crea un generador de valores en posiciones pares con el type válido
    valores_pos_pares = (n for n in valores[::2] if es_int_o_float(n))

    # calculos
    for n in valores_pos_pares:
      total *= n

    return total

  else:
    print("Se espera una lista")


# uso de la función
valores = [2, 3, -1.5, 4,-1, 'a', 1.5, 'hola']

# invocando la opción 1
print(multi_pos_pares(valores))

# invocando la opción 2
print(multi_pos_pares_v2(valores))
```

```python
# se define la función
def info_rectangulo(a, b):
  "Recibe los lados de un rectangulo y retorna (Area, Perímetro)"

  A = round(a * b, 1)
  P = round(2 * (a + b), 1)
  
  # se retorna como tupla (packing) 
  return A, P  # parentesis de la tupla implícitos, se asume tupla por la coma


# se invoca asignando como tupla (sin unpacking) 
result = info_rectangulo(3.2, 8.5)
print("Como tupla: {}".format(result)) # (27.2, 23.4)

# con unpacking
area, perimetro = info_rectangulo(7, 5)
print("Area: {}, Perimetro: {}".format(area, perimetro))

```

```python
# se define la función con hint para recibir el detalle como diccionario
def calcular_total_compra(detalle_compra: dict):
  "Recibe un diccionario con los items y cantidades y devuelve el importe total"

  # lista de precios
  precios = {
      'Fideos': 188,
      'Azucar': 83,
      'Sal': 71,
      'Huevos': 111,
      'Crema': 179,
      'Gaseosa': 193,
      'Pan': 93,
      'Salsa': 123
  }

  importe_total = 0

  # se itera sobre el diccionario del detalle, cada item es la clave
  for item in detalle_compra:

    # se valida la key en el diccionario
    if item in precios:

      # calcula y acumula
      importe_total += precios[item] * detalle_compra[item]
    else:
      print("Precio del producto '{}' desconocido".format(item))
  
  return importe_total
```

````python
# se usa la función

compra1 = {"Fideos": 2, "Papas":2, "Pan": 2, "Salsa": 2, "Gaseosa": 2, "Carne": 1} 

print(calcular_total_compra(detalle_compra=compra1))
````

Crear una función que reciba un texto y devuelva un diccionario con la frecuencia de cada palabra en el mismo.


````python
# Espacio para resolución del Ejercicio 9




# Fin resolución del ejercicio 9
````

##### **Ver resolución Ejercicio 9**

```python
# se importa librería para expresiones regulares
import re

def contar_palabras(texto: str):
  "Recibe un texto y retorna un diccionario con la frecuencia de cada palabra"

  # se pasa todo a minúsculas y se eliminan espacios al inicio o fin
  texto = texto.lower().strip()

  # se descartan caracteres de puntuación 
  texto = re.sub(r'[^\w\s]|\n','', texto)

  # se crea diccionario de frecuencias con las palabras como clave inicializado
  # con 0 en el valor donde se va a indicar las ocurrencias de cada una 
  frecuencias = dict.fromkeys(texto.split(" "), 0)

  # se itera sobre las claves del diccionario de frecuencias 
  for palabra in frecuencias:
    
    # se asigna la cantidad a esa clave del diccionario
    frecuencias[palabra] = texto.count(palabra)
  
  return frecuencias
```
